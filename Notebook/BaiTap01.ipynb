```python
def tach_diem(chuoi):
    l = chuoi.split('  ')
    mon = []
    diem = []
    for i in range(0,len(l),2):
        mon.append(l[i].strip(':')) 
        diem.append(float(l[i+1]))
        
    return pd.DataFrame({'mon':mon,'diem':diem})
```

```python
tach_diem(df.iloc[0]['DIEM_THI']).mean() # 1 thi sinh
```

```python
df['diem_trung_binh'] = df['DIEM_THI'].map(lambda x: tach_diem(x)['diem'].mean())
#df['diem_trung_binh'] = df.['DIEM_THI'].apply
```

```python
df[['SOBAODANH','diem_trung_binh']].to_csv('tuyen_sinh_dtb.csv',encoding='utf-8',index=False)
```
