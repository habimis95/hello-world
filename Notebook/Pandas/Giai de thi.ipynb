```python
# Viết hàm đọc tập tin csv và trả về dataframe đọc được
def read_data_csv(ten_tap_tin):
    df = pd.read_csv(ten_tap_tin, index_col=0)
    return df
```

```python
# đọc các sinh viên khoa TO
df_to = read_data_csv('sinh_vien_TO.csv')
# đọc các sinh viên khoa VL
df_vl = read_data_csv('sinh_vien_VL.csv')
```
